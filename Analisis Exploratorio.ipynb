## Las Provincias con Promedio de Años de Antiguedad Mas Alto

# ¿Como es el precio de cada Tipo de Propiedad dependiendo de la cantidad de Habitaciones que tenga?

```python
for_heatmap = df.pivot_table(index='tipodepropiedad', columns='habitaciones', values='precio', aggfunc='mean')
for_heatmap
for_heatmap = for_heatmap.iloc[[True, False, True, True,True,False,True,True,False,False,True,True,False,False,True,False,False,True,True,False,False,True],:6]
for_heatmap
```

```python
g = sns.heatmap(for_heatmap,cmap="BuPu", linewidths=.5)
g.set_title("El Precio Promedio de cada tipo de Propiedad por Cantidad de Habitaciones", fontsize=11)
g.set_xlabel("Cantidad de Habitaciones",fontsize=11)
g.set_ylabel("Tipo de Propiedades", fontsize=11)
```

# ¿Como es el precio de cada Tipo de Propiedad dependiendo de la cantidad de Baños que tenga?

```python

g = sns.heatmap(for_heatmap, cmap="YlGnBu", linewidths=.5)
g.set_title("El Precio Promedio de cada tipo de Propiedad por Cantidad de Baños", fontsize=11)
g.set_xlabel("Cantidad de Baños",fontsize=11)
g.set_ylabel("Tipo de Propiedades", fontsize=11)
```
