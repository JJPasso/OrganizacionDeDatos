## Las Provincias con Promedio de Años de Antiguedad Mas Alto

```python
prom_ant_prov = df.loc[:,['provincia','antiguedad']].groupby("provincia").\
    agg([np.mean,np.size])
show = prom_ant_prov.loc[prom_ant_prov[('antiguedad','size')]>5000,:].sort_values(('antiguedad','mean'), ascending=False)
show
```

```python
show = show.iloc[:10,:]
g = sns.barplot(x=show['antiguedad','mean'].values, y=show['antiguedad','mean'].index, orient='h')
g.set_title("Las Provincias con Promedio de Años de Antiguedad Mas Altos", fontsize=15)
g.set_xlabel("Promedio de Años de Antiguedad", fontsize=12)
g.set_ylabel("Provincias", fontsize=12)
```
